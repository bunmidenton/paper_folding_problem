```python
folded_list = [0.00008,]
thickness = 0.00008
for i in range(43):
    thickness = thickness * 2
    folded_list.append(thickness)
    print(folded_list)
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_list)
plt.show()
```
